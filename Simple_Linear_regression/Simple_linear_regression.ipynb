# Importan Librerias 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

# Importar Dataseet

```python
dataseet = pd.read_csv("Salary_Data.csv")
```

```python
print(dataseet.head(10))
```

```python
X = dataseet.iloc[:,:-1].values
```

```python
print(X[0:10])
```

```python
Y = dataseet.iloc[:, 1].values
```

```python
print(Y[0:10])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test =train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print("Intercepto (b0):", regressor.intercept_)
print("Pendiente (b1):", regressor.coef_[0])
print("R^2 Score:", r2_score(y_test, y_pred))
# Error Cuadrático Medio (MSE)
print("MSE:", mean_squared_error(y_test, y_pred))
# Raíz del Error Cuadrático Medio (RMSE)
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
# Gráfica de ENTRENAMIENTO
plt.scatter(x_train, y_train, color='red', label='Datos reales')
plt.plot(x_train, regressor.predict(x_train), color='blue', label='Línea de regresión')
plt.title('Salario vs Experiencia (Entrenamiento)')
plt.xlabel('Años de experiencia')
plt.ylabel('Salario')
plt.legend()
plt.show()

# Gráfica de PRUEBA
plt.scatter(x_test, y_test, color='red', label='Datos reales')
plt.plot(x_train, regressor.predict(x_train), color='blue', label='Línea de regresión')
plt.title('Salario vs Experiencia (Prueba)')
plt.xlabel('Años de experiencia')
plt.ylabel('Salario')
plt.legend()
plt.show()
```
